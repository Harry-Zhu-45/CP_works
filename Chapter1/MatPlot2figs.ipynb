```python
from pylab import *
%matplotlib inline


Xmin = -5.0
Xmax = 5.0
Npoints = 500
DelX = (Xmax-Xmin) / Npoints

x1 = arange(Xmin, Xmax, DelX)       # x1 range
x2 = arange(Xmin, Xmax, DelX/20)    # Different x2 range
y1 = -sin(x1)*cos(x1*x1)            # Function 1
y2 = exp(-x2/4)*sin(x2)             # Function  2
print("\n Now plotting, look for Figures 1 & 2 on desktop")

# Figure 1
figure(1)

# # 1st subplot in figure
subplot(2, 1, 1)
plot(x1, y1, 'r', lw=2)
xlabel('x')
ylabel('f(x)')
title('-sin(x)*cos(x^2)')
grid(True)

# # 2nd subplot in first figure
subplot(2, 1, 2)
plot(x2, y2, '-', lw=2)
xlabel('x')
ylabel('f(x)')
title('exp(-x/4)*sin(x)')


# Figure 2
figure(2)

# # 1st subplot in second figure
subplot(2, 1, 1)
plot(x1, y1*y1, 'r', lw=2)
xlabel('x')
ylabel('f(x)')
title('exp(-x/2)*sin^2(x)')

# # 2nd subplot in second figure
subplot(2, 1, 2)
plot(x2, y2*y2, '-', lw=2)
xlabel('x')
ylabel('f(x)')
title('exp(-x/2)*sin^2(x)')
grid(True)


plt.tight_layout()  # matplotlib's auto optimization
show()
```
