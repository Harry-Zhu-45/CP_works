```python
import numpy as np
import matplotlib.pyplot as plt


xdata = np.array([14.4, 14.41, 14.51, 14.625, 14.675, 15.33, 15.41, 15.43, 15.8, 16.415,
                  16.47, 16.57, 16.64, 16.651, 17.065, 17.63, 17.88, 17.975, 18.015, 18.035,
                  18.935, 19.051])

ydata = np.array([706.52, 708.19, 690.75, 671.64, 667.81, 587.56, 579.07, 576.96, 546.07, 504.77,
                  501.57, 496.03, 492.19, 491.61, 471.31, 447.15, 438.79, 435.84, 434.75, 433.92,
                  407.78, 404.66])

func1 = np.polyfit(xdata, ydata, 4)
coefficient1 = np.poly1d(func1)
print(coefficient1)
```

```python
# Plot the data
fig = plt.figure()
plt.title('$y = 0.5075 x^2 - 36.59 x^3 + 994 x^2 - 1.21e4 x + 5.618e4$')
plt.xlabel('Reading of drum: $\mathrm{mm}$')
plt.ylabel('spectrum: $\mathrm{nm}$')


plt.scatter(xdata, ydata, label='Data', color='red', marker='o')
plt.plot(xdata, coefficient1(xdata), label='polynomial')

# show text: coefficient
# plt.text(20, 20, '$y = 0.5075 x^2 - 36.59 x^3 + 994 x^2 - 1.21e4 x + 5.618e4$')
plt.legend(loc='best')
plt.show()
```
