```python
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('test 2*2 axes')


# subplot 1
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

ax1.set_title('f(x)=exp(-x)*cos(2pi x)')
ax1.plot(t1, f(t1), color='tab:blue', marker='o')
ax1.plot(t2, f(t2), color='black')


# subplot 2
x = np.linspace(-4, 4, 6)
y = np.linspace(-4, 4, 6)
X, Y = np.meshgrid(x, y)
U = X + Y
V = Y - X

ax2.set_title('plt.quiver')
ax2.quiver(X, Y, U, V, color="C0", angles='xy', scale_units='xy', scale=5, width=.015)
ax2.set(xlim=(-5, 5), ylim=(-5, 5))


# subplot 3
x = np.arange(-2, 2, 0.2)
y = np.arange(-2, 2, 0.25)
X, Y = np.meshgrid(x, y)
Z = X*np.exp(-X**2-Y**2)
V, U = np.gradient(Z, 0.2, 0.2)

ax3.set_title('line of force')
ax3.scatter(X, Y, color="blue", s=0.05)
ax3.quiver(X, Y, U, V, color="red", pivot="tip", units="inches")


# subplot 4
n = 2000000  # number of points
xdata = np.random.random(n)
ydata = np.random.random(n)

index = xdata**2 + ydata**2 < 1
count = np.sum(index)
count_pi = 4 * (count / n)

x_pi = xdata[index]
y_pi = ydata[index]
x_not_pi = xdata[~index]
y_not_pi = ydata[~index]

ax4.axis('equal')
ax4.scatter(x_pi, y_pi, color="blue", s=1/np.sqrt(n))
ax4.scatter(x_not_pi, y_not_pi, color="orange", s=1/np.sqrt(n))
ax4.set_title("Monte Carlo pi = %8.6f\nwhile n = %2.0f" % (count_pi, n))


plt.tight_layout()
plt.show()
```
