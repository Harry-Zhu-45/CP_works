```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots()
ax.set_title('Grade Inflation')  # Title and labels
ax.set_xlabel('Years in College')
ax.set_ylabel('GPA')


xa = np.array([-1, 5])  # For horizontal line
ya = np.array([0, 0])   # " "
ax.plot(xa, ya)         # Draw horizontal line


# Data set 0 points
x0 = np.array([0, 1, 2, 3, 4])
y0 = np.array([-1.4, 1.1, 2.2, 3.3, 4.0])
ax.plot(x0, y0, 'bo')   # Data set 0 = blue circles
ax.plot(x0, y0, 'g')    # Data set 0 = line


# Data set 1 points
x1 = np.arange(0, 5, 1)
y1 = np.array([4.0, 2.7, -1.8, -0.9, 2.6])
ax.plot(x1, y1, 'r')


errTop = np.array([1.0, 0.3, 1.2, 0., 0.1])     # Asymmetric error bars
errBot = np.array([2.0, 0.6, 2.3, 1.8, 0.4])
ax.errorbar(x1, y1, [errBot, errTop], fmt='o')  # Plot error bars


ax.grid(True)
plt.show()
```
