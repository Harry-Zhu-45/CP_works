```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# plt.style.use('default')
# mpl.rcParams['text.usetex'] = True  # LaTeX in text
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']

Xmin = -5.0
Xmax = 5.0
Npoints = 500
DelX = (Xmax-Xmin) / Npoints


x1 = np.arange(Xmin, Xmax, DelX)       # x1 range
x2 = np.arange(Xmin, Xmax, DelX/20)    # Different x2 range
y1 = -np.sin(x1) * np.cos(x1*x1)       # Function 1
y2 = np.exp(-x2/4) * np.sin(x2)        # Function  2
```

```python
# Figure 1
fig1, [ax1, ax2] = plt.subplots(2, 1)

# 1st subplot in Figure 1
ax1.plot(x1, y1, 'r', lw=2)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title('-sin(x)*cos(x^2)')
ax1.grid(True)

# 2nd subplot in Figure 1
ax2.plot(x2, y2, '-', lw=2)
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.set_title('exp(-x/4)*sin(x)')

# plt.tight_layout()  # matplotlib's auto optimization
fig1.show()
```

```python
# Figure 2
fig2, [ax1, ax2] = plt.subplots(2, 1)

# # 1st subplot in the second figure
ax1.plot(x1, y1*y1, 'r', lw=2)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title('exp(-x/2)*sin^2(x)')

# # 2nd subplot in the second figure
ax2.plot(x2, y2*y2, '-', lw=2)
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.set_title('exp(-x/2)*sin^2(x)')
ax2.grid(True)

# plt.tight_layout()  # matplotlib's auto optimization
plt.show()
```
