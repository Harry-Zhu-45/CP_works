## Bisection 二分法

$$
f(a) * f(b) < 0, \quad a < b
$$

可以用二分法计算函数 $f(x)$ 在 $\in[a,b]$ 之间的零点

每次在中间取点计算函数值，与两端的函数值比较，零点应该在函数值异号的两点之间。

### fixed point

不动点

$$
g(p) = p
$$

> 寻找不动点的问题可以转化为了寻找零点的问题。

$$
g \in [a, b], \quad g(x) \in [a, b] \tag{1}
$$

则至少存在一个不动点

$$
|g^{\prime}(x)| \le k \tag{2}
$$

则只存在一个不动点。

一开始选取一个点 $P_0$，然后迭代

$$
P_n = g(P_{n-1})
$$

$$
P = \lim_{n \to \infty} P_n = \lim_{n \to \infty} g(P_{n-1}) = g(\lim_{n \to \infty} P_{n-1}) = g(P)
$$

就得到了不动点 $P$

## Weierstrass 逼近定理

对任意 $\varepsilon > 0$，存在多项式函数 $p(x)$，使得

$$
|f(x)-p(x)| < \varepsilon, \quad x \in [a, b]
$$

## 拉格朗日插值

给定几个分立的数据点 $\{x_i\}$，使用多项式函数 $P(x)$ 拟合函数 $f(x)$

> 多项式函数需要穿过每个给定的数据点

$$
P_n(x) = \sum_{i=0}^{n} l_{n,i}(x) f(x_i)
$$

$x=x_i$ 时，$P_n(x_i)=f(x_i)$，所以

$$
l_{n,i}(x) =
\begin{cases}
1, & x = x_i \\
0, & x \ne x_i
\end{cases}
$$

不难想到满足 $l_{n,i}=0$ 的形式，$\prod_{x \ne x_i} (x-x_i)$，$l_i=1$ 的条件相当于是归一化

$$
l_{n,i}(x) = \frac{(x-x_1)(x-x_2) \cdots (x-x_n)}{(x_i-x_1)(x_i-x_2) \cdots (x_i-x_n)} = \frac{\prod_{j=1,j\ne i}^{n} (x-x_j)}{\prod_{j=1,j\ne i}^{n} (x_i-x_j)}
$$

考虑多项式函数与原函数的误差

> Numerical Analysis Page 112

$$
f(x) = P(x) + \frac{f^{(n+1)}(\xi(x))}{(n+1)!} (x-x_0)(x-x_1) \cdots (x-x_n)
$$

# 偏微分方程

波动方程 薛定谔方程

麦克斯韦方程组 推导 波动方程

$$
\nabla \times \vec{E} = -\mu \frac{\partial \vec{H}}{\partial t}
$$

$$
\nabla \times \vec{H} = \varepsilon \frac{\partial \vec{E}}{\partial t} + \vec{J}(t)
$$

$$
\nabla \cdot (\varepsilon \vec{E}) = \rho(t)
$$

$$
\nabla \cdot (\mu \vec{E}) = 0
$$

$$
\vec{X}(t) = \sqrt{\mu} \vec{H}(t), \quad \vec{Y}(t) = \sqrt{\varepsilon} \vec{E}(t)
$$

改写为

$$
\frac{\partial}{\partial t}
\begin{pmatrix}
\vec{X}(t) \\
\vec{Y}(t)
\end{pmatrix}
=
\begin{pmatrix}
0 & -\frac{1}{\sqrt{\mu}} \nabla \times \frac{1}{\sqrt{\varepsilon}} \\
\frac{1}{\sqrt{\varepsilon}} \nabla \times \frac{1}{\sqrt{\mu}} & 0
\end{pmatrix}
\begin{pmatrix}
\vec{X}(t) \\
\vec{Y}(t)
\end{pmatrix}
$$

$$
H^T = -H
$$

麦克斯韦方程组相当于

$$
\frac{\partial}{\partial t} \Psi(t) = H \Psi(t)
$$

如果 H 不含时间，偏微分方程的解可以写为

$$
\Psi(t) = e^{tH} \Psi(0) \equiv u(t) \Psi(0)
$$

$$
u^T(t) = u(-t) = u^{-1}(t) = e^{-tH}
$$

时间演化的电磁场

一维：

$$
\psi(i, t) =
\begin{cases}
X_y(i, t) = \sqrt{\mu_i} H_y(i, t) \quad odd \\
Y_z(i, t) = \sqrt{\varepsilon_i} E_z(i, t) \quad even
\end{cases}
$$

$$
\frac{\partial }{\partial t} \psi = H \psi
$$

$$
H =
\begin{pmatrix}
0 & \beta_{21} \\
-\beta_{21} & 0 & \beta_{23} \\
 & -\beta_{23} & 0 \\
 & & & \ddots
\end{pmatrix}
$$

$$
\beta_{ij} = \frac{1}{\delta \sqrt{\varepsilon_i \mu_i}}
$$

$$
e^{\alpha\begin{pmatrix}0&1\\-1&0\end{pmatrix}}
\begin{pmatrix}
\psi_i \\
\psi_j
\end{pmatrix}
=
\begin{pmatrix}
\cos \alpha & \sin \alpha \\
-\sin \alpha & \cos \alpha
\end{pmatrix}
\begin{pmatrix}
\psi_i \\
\psi_j
\end{pmatrix}
$$

二维

$$
\psi_{TM}(t) =
\begin{pmatrix}
X_x(x,y,t) \\
X_y(x,y,t) \\
Y_z(x,y,t)
\end{pmatrix}
$$

$$
\frac{\partial}{\partial t} \psi_{TM} = H \psi_{TM}
$$

$$
H =
\begin{pmatrix}
0 & 0 & -\frac{1}{\sqrt{\mu}} \frac{\partial}{\partial y} \frac{1}{\sqrt{\varepsilon}} \\
0 & 0 & \frac{1}{\sqrt{\mu}} \frac{\partial}{\partial x} \frac{1}{\sqrt{\varepsilon}} \\
-\frac{1}{\sqrt{\varepsilon}} \frac{\partial}{\partial y} \frac{1}{\sqrt{\mu}} & \frac{1}{\sqrt{\varepsilon}} \frac{\partial}{\partial x} \frac{1}{\sqrt{\mu}} & 0
\end{pmatrix}
$$
