## Bisection 二分法

$$
f(a) * f(b) < 0, \quad a < b
$$

可以用二分法计算函数 $f(x)$ 在 $\in[a,b]$ 之间的零点

每次在中间取点计算函数值，与两端的函数值比较，零点应该在函数值异号的两点之间。

### fixed point

不动点

$$
g(p) = p
$$

> 寻找不动点的问题可以转化为了寻找零点的问题。

---

$$
g \in [a, b], \quad g(x) \in [a, b] \tag{1}
$$

则至少存在一个不动点

$$
|g^{\prime}(x)| \le k \tag{2}
$$

则只存在一个不动点。

一开始选取一个点 $P_0$，然后迭代

$$
P_n = g(P_{n-1})
$$

$$
P = \lim_{n \to \infty} P_n = \lim_{n \to \infty} g(P_{n-1}) = g(\lim_{n \to \infty} P_{n-1}) = g(P)
$$

就得到了不动点 $P$

## 拉格朗日插值

给定几个分立的数据点 $\{x_i\}$，使用多项式函数 $P(x)$ 拟合函数 $f(x)$

> 多项式函数需要穿过每个给定的数据点

$$
P_n(x) = \sum_{i=0}^{n} l_{n,i}(x) f(x_i)
$$

$x=x_i$ 时，$P_n(x_i)=f(x_i)$，所以

$$
l_{n,i}(x) =
\begin{cases}
1, & x = x_i \\
0, & x \ne x_i
\end{cases}
$$

不难想到满足 $l_{n,i}=0$ 的形式，$\prod_{x \ne x_i} (x-x_i)$，$l_i=1$ 的条件相当于是归一化

$$
l_{n,i}(x) = \frac{(x-x_1)(x-x_2) \cdots (x-x_n)}{(x_i-x_1)(x_i-x_2) \cdots (x_i-x_n)} = \frac{\prod_{j=1,j\ne i}^{n} (x-x_j)}{\prod_{j=1,j\ne i}^{n} (x_i-x_j)}
$$

考虑多项式函数与原函数的误差

> Numerical Analysis Page 112

$$
f(x) = P(x) + \frac{f^{(n+1)}(\xi(x))}{(n+1)!} (x-x_0)(x-x_1) \cdots (x-x_n)
$$

## 数值 积分/微分

$$
\begin{aligned}
\int_{a}^{b} f(x) dx &= \lim_{max \Delta x\to 0} \sum_{i=1}^{n} f(x_i) \Delta x_i \\
&= \lim_{n\to\infty} \frac{b-a}{n} \sum_{i=1}^{n} f(x_i)
\end{aligned}
$$

### 有限差分法常用算法

- Verlet 算法

$$
v(t) = \frac{r(t+\frac{1}{2}\delta t)-r(t-\frac{1}{2}\delta t)}{\delta t}
$$

$$
a(t) = \frac{v(t)-v(t-\delta t)}{\delta t}
$$

$$
r(t+\delta t) = 2 r(t) - r(t-\delta t) + \delta t^2 a(t)
$$

- Velocity-Verlet 算法 (直接给出下个时步的位置、速度、加速度)

$$
r(t+\delta t) = r(t) + v(t) \delta t + \frac{1}{2} a(t) \delta t^2
$$

$$
v(t+\delta t) = v(t) + \frac{1}{2} [a(t) + a(t+\delta t)] \delta t
$$

$$
a(t+\delta t) = \frac{v(t+\delta t)-v(t)}{\delta t}
$$
