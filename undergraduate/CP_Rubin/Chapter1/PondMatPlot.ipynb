```python
import matplotlib.pyplot as plt
import numpy as np


def fx(x):
    return x * np.power(np.sin(x), 2)


N = 1000
x1 = np.arange(0, 2 * np.pi + 2 * np.pi / N, 2 * np.pi / N)
y1 = fx(x1)


# plot init
fig, ax = plt.subplots()
ax.set_xlabel("x", fontsize=20)
ax.set_ylabel(r"$f(x)=x\,\sin^2 (x)$", fontsize=20)
ax.set_xlim((0, 2 * np.pi))
ax.set_ylim((0, 5))
ax.set_xticks([0, np.pi, 2 * np.pi])
ax.set_xticklabels([r"0", r"$\pi$", r"$2\pi$"])

ax.plot(x1, y1, "c", linewidth=2)


Npts = 3000000
analyt = np.pi**2

rng = np.random.default_rng()
xx = 2.0 * np.pi * rng.random(Npts)  # 0 =< x <= 2pi
yy = 5 * rng.random(Npts)  # 0 =< y <= 5

xi = xx[yy <= fx(xx)]
yi = yy[yy <= fx(xx)]
xo = xx[yy > fx(xx)]
yo = yy[yy > fx(xx)]

j = len(xi)

boxarea = 2 * np.pi * 5  # Box area
area = boxarea * j / (Npts - 1)  # Area under curve


# plot
ax.plot(xo[:100], yo[:100], "bo", markersize=3)
ax.plot(xi[:100], yi[:100], "ro", markersize=3)
ax.set_title("Answers: Analytic = %5.3f, MC = %5.3f" % (analyt, area))

plt.show()
```
