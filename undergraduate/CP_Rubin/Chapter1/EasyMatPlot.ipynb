```python
import numpy as np
import matplotlib.pyplot as plt


Xmin = -5.0
Xmax = +5.0
Npoints = 500
DelX = (Xmax - Xmin) / Npoints


x = np.arange(Xmin, Xmax, DelX)
y = np.sin(x) * np.sin(x*x)  # function of x array


print('arange => x[0], x[1], x[499] = %8.2f %8.2f %8.2f' % (x[0], x[1], x[499]))
print('arange => y[0], y[1], y[499] = %8.2f %8.2f %8.2f' % (y[0], y[1], y[499]))
print("\n Now doing the plotting thing , look for Figure1 on desktop")


fig = plt.figure()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) vs x')
plt.text(-1.75, 0.75, 'MatPlotLib \n Example')
plt.plot(x, y, '-', lw=2)
plt.grid(True)


plt.show()
```
