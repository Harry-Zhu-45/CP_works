```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin

# Initialization
a = 0.0
b = 10.0
n = 100

y = np.zeros(2, float)
y[0] = 3.0
y[1] = -5.0

t = a
h = (b - a) / n


def f(t, y):  # Force function
    f_return = np.zeros(2, float)
    f_return[0] = y[1]
    f_return[1] = -100.0 * y[0] - 2.0 * y[1] + 10.0 * sin(3.0 * t)
    return f_return


def rk4(t, y, h):
    k1 = h * f(t, y)
    k2 = h * f(t + h / 2.0, y + k1 / 2.0)
    k3 = h * f(t + h / 2.0, y + k2 / 2.0)
    k4 = h * f(t + h, y + k3)

    y_new = y + (k1 + 2.0 * k2 + 2.0 * k3 + k4) / 6.0
    return y_new


# Save results for plotting
t_values = [t]
y0_values = [y[0]]
y1_values = [y[1]]

while t < b:  # Time loop
    if (t + h) > b:
        h = b - t  # Last step

    y = rk4(t, y, h)
    t = t + h

    t_values.append(t)
    y0_values.append(y[0])
    y1_values.append(y[1])

# Plot with matplotlib
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(t_values, y0_values, label="y[0]")
plt.title("RK4 Solution")
plt.xlabel("t")
plt.ylabel("y[0]")
plt.xlim(0, 10)
plt.ylim(-2, 3)
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t_values, y1_values, label="y[1]")
plt.xlabel("t")
plt.ylabel("y[1]")
plt.xlim(0, 10)
plt.ylim(-25, 18)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
