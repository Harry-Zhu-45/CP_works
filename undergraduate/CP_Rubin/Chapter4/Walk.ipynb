# Random Walk

We will present the simplest approach for a 2D walk, with a minimum of theory, and end up with a model for
normal diffusion.


```python
import matplotlib.pyplot as plt
import numpy as np


jmax = int(1e4)

rng = np.random.default_rng()
xlist = np.cumsum(2 * rng.random(jmax) - 1)
ylist = np.cumsum(2 * rng.random(jmax) - 1)

plt.axis("equal")
plt.plot(xlist, ylist, "y.-", markersize=0.01 / np.sqrt(jmax))
plt.show()
```
