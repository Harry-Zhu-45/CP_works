```python
import matplotlib.pylab as plt
import numpy as np


para_lambda = 0.01  # Decay constant
time_max = 1600
seed = 68111  # Params
nloop = max = 1000000  # Initial value


t = np.arange(0, time_max)
y = np.zeros(time_max)

rng = np.random.default_rng()
for time in range(0, time_max):
    nloop -= np.sum(rng.random(nloop) < para_lambda)
    y[time] = nloop


fig = plt.figure()
plt.title("Spontaneous Decay")
plt.xlabel("Time")
plt.ylabel("Number Left")
plt.yscale("log")

plt.plot(t, y)

plt.tight_layout()
plt.show()
```
